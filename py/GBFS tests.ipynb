```python
import json
import os
import pandas as pd
```

```python
path_to_json = '../.data/gbfs/lime'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
json_objs = []
for file in json_files:
    with open(path_to_json + '/' + file) as f:
        json_objs.append(json.load(f))
gbfs = []
for r in json_objs:
    for b in r['data']['bikes']:
        x = b
        x['provider'] = 'Lime'
        x['update_time'] = r['last_updated']
        gbfs.append(x)
```

```python
path_to_json = '../.data/gbfs/spin'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
json_objs = []
for file in json_files:
    with open(path_to_json + '/' + file) as f:
        json_objs.append(json.load(f))

for r in json_objs:
    for b in r['data']['bikes']:
        x = b
        x['provider'] = 'Spin'
        x['update_time'] = r['last_updated']
        gbfs.append(x)
```

```python
path_to_json = '../.data/gbfs/clevr'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
json_objs = []
for file in json_files:
    with open(path_to_json + '/' + file) as f:
        json_objs.append(json.load(f))
gbfs = []
for r in json_objs:
    for b in r['data']['bikes']:
        x = b
        x['provider'] = 'Clevr'
        x['update_time'] = r['last_updated']
        gbfs.append(x)
```

```python
gbfs = pd.DataFrame.from_records(gbfs)
gbfs.head()
```

```python
from datetime import datetime
from pytz import timezone
gbfs['update_time'] = gbfs.update_time.map(lambda t: datetime.fromtimestamp(t).astimezone(timezone('US/Pacific')))
gbfs.head()
```

```python
gbfs['update_hour'] = gbfs.update_time.map(lambda t: t.hour)
gbfs['update_minute'] = gbfs.update_time.map(lambda t: t.minute)
gbfs.head()
```

```python
lime = gbfs[gbfs.provider == 'Lime']
lime = lime[lime.update_hour == 10]
lime = lime[lime.update_minute < 50]
spin = gbfs[gbfs.provider == 'Spin']
spin = spin[spin.update_hour == 11]
spin = spin[spin.update_minute < 50]
jump = gbfs[gbfs.provider == 'Jump']
jump = jump[jump.update_hour == 15]
jump = jump[jump.update_minute < 50]
```
