# Voronori diagram of street intersections
## Import intersections to geopandas data frame

```python
import geopandas as gpd
intersections = gpd.read_file("data/streets.gdb", driver='FileGDB', layer='STREET_INTERSECTION')
intersections.head()
```

```python
intersections.geometry.crs
```

```python
import geoplot
geoplot.pointplot(intersections)
```

```python
cities = gpd.read_file("https://opendata.arcgis.com/datasets/1559e31273654eb9858397861f1fdefa_10.geojson")
geoplot.polyplot(cities, projection=geoplot.crs.Mercator())
```

```python
cities.head()
```

```python
geoplot.pointplot(intersections)
```

```python
geoplot.polyplot(cities)
```

```python
intersections = intersections.to_crs(epsg='4326')
ax = cities.plot(cmap='Purples')
intersections.plot(ax=ax)
```
